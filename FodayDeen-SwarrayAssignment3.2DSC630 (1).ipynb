# Los Angeles Dodgers Attendance Analysis
## Step 1: Data Understanding and Preparation
### Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Load Data and display first few rows and columns

```python
dodgers_data = pd.read_csv("C:\\Users\\lahai\\Downloads\\dodgers-2022.csv")
dodgers_data.head()
```

# Data Cleaning
### Check for missing values and drop or fill missing values when necessary

```python
dodgers_data.isnull().sum()
```

```python
dodgers_data = dodgers_data.dropna()
```

```python
dodgers_data ['game_date'] = pd.to_datetime(dodgers_data['month'].astype(str)+ '-' + dodgers_data['day'].astype(str) + '-2023')
print("\nUpdated DataFrame with Combined Date:")
print(dodgers_data)
```

```python
dodgers_data['game_date'] = pd.to_datetime(dodgers_data['game_date'])
dodgers_data.dtypes
```

```python
# Plot overall attendance trends over the season
plt.figure(figsize=(12, 6))
dodgers_data.groupby('game_date')['attend'].sum().plot()
plt.title('Attendance Trends Over the Season')
plt.xlabel('Date')
plt.ylabel('Attendance')
plt.xticks(rotation=45)
plt.show()
```

```python
# Compare attendance against different opponents
plt.figure(figsize=(12, 6))
sns.boxplot(x='opponent', y='attend', data=dodgers_data)
plt.title('Attendance by Opponent')
plt.xticks(rotation=45)
plt.show()
```

```python
dodgers_data['day_of_week'] = dodgers_data['game_date'].dt.day_name()
```

```python
# Day of the Week Impart
plt.figure(figsize=(12, 6))
sns.boxplot(x='day_of_week', y='attend', data=dodgers_data, order=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])
plt.title('Attendance by Day of the Week')
plt.show()
```

```python
# Promotions and Event
plt.figure(figsize=(12, 6))
sns.boxplot(x='fireworks', y='attend', data=dodgers_data)
plt.title('Attendance on Promotional vs Non-Promotional Nights')
plt.xticks([0, 1], ['Non-Promotional','Promotional'])
plt.show()
```

```python
# Weather Conditions
plt.figure(figsize=(12, 6))
sns.boxplot(x='temp', y='attend', data=dodgers_data)
plt.title('Attendance by Weather Conditions')
plt.show()
```

```python
print("Key Observations:")
print("1. Attendance peaks during week.")
print("2. Promotional nights significantly increase attendance.")
print("3. Rival games draw larger crowds.")
```
