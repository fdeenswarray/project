```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Simulating a dataset for childcare prices
data = {'Location': ['California','Texas','New York','Florida','Illinois','Ohio','Georgia'],'Average_price':[1200, 900, 1500, 1000, 950, 850, 800],
        'Type_of_Care':['Daycare','Family Care','Preschool','Daycare','Family Care','Preschool', 'Daycare'], 
        'Age_Group':['Infants','Toddlers','Preschoolers','Infants','Toddlers','Preschoolers','Infants']}
df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='Type_of_Care', y='Average_price', data=df, palette='viridis')
plt.title('Average Childcare Prices by Type of Care')
plt.xlabel('Type of Care')
plt.ylabel('Average Price ($)')
plt.show()
```

```python
# Pie chart showing distribution of costs across different age groups
age_group_data = df['Age_Group'].value_counts()
plt.figure(figsize=(8, 5))
plt.pie(age_group_data, labels=age_group_data.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Distribution of Childcare Costs by Age Group')
plt.axis('equal')
plt.show()
```

```python
# Heatmap example
heatmap_data = pd.pivot_table(df, values='Average_price', index='Location', columns='Type_of_Care')

plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, cmap='coolwarm',fmt=".0f")
plt.title('Childcare Prices Heatmap')
plt.ylabel('Location')
plt.xlabel('Type of Care')
plt.show()
```

# Design Narrative

Collectively, these decisions were driven by the need to balance clarity, interactivity, and relatability for parents, providers, and policymakers.
